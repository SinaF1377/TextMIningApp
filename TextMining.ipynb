# Text Mining

```python
import pandas as pd
```

```python
data = pd.read_csv('per.csv', encoding='utf8')
# encoding :: این پارت زمانی استفاده میشود که فایل ما فارسی است
data.head()
```

```python
# stopwords :: کلماتی که از نظر معنای معنای خاصی ندارند و در یادگیری کمکی به ما نمیکنند
with open('stopwords.txt', encoding='utf8') as stopwords_file:
    stopwords = stopwords_file.readlines()
stopwords = [line.replace('\n', '')  for line in stopwords]
stopwords
```

```python
import nltk 
nltk.download('stopwords')
```

```python
nltk_stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(nltk_stopwords)
stopwords
```

```python
import hazm
stemmer = hazm.Stemmer()
```

```python
stemmer.stem('خانه هایشان')
```

```python
from hazm import word_tokenize
```

```python
# make data fram with two part :: data(title + body) and category
dataset = pd.DataFrame(columns=('title_body', 'category'))
for index , row in data.iterrows():
    # title and body
    title_body = row['Title'] + ' ' + row['Body']
    # tokenize all data
    title_body_tokenized = word_tokenize(title_body)
    # delete stopwords
    title_body_tokenized_filtered = [w for w in title_body_tokenized if not w in stopwords]
    # save data in datafram with category
    title_body_tokenized_filtered_stemed = [stemmer.stem(w) for w in title_body_tokenized_filtered]

    dataset.loc[index] = {
        'title_body' : ' '.join(title_body_tokenized_filtered_stemed),
        'category' : row['Category2'].replace('\n', ' ')
    }
```

```python
dataset.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorize = TfidfVectorizer()
vectorize.fit(dataset['title_body'])
```

```python
import pickle
with open('News.jsdh', 'wb') as f:
    pickle.dump(vectorize, f)
```

```python
X = vectorize.transform(dataset['title_body'])
X
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(dataset['category'])
```

```python
le.classes_
```

```python
with open('Leabel.jsdh', 'wb') as f:
    pickle.dump(le, f)
```

```python
import numpy as np
np.unique(dataset['category'])
```

```python
print(np.shape(X) + np.shape(y))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test , y_train, y_test = train_test_split(X, y)
```

```python
from sklearn import svm
```

```python
svmc = svm.SVC()
svmc.fit(X_train, y_train)
```

```python
with open('svc.jsdh', 'wb') as f:
    pickle.dump(svmc, f)
```

```python
svmc.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred = svmc.predict(X_test)
```

```python
print(classification_report(y_pred, y_test))
```

```python
print(confusion_matrix(y_pred, y_test))
```
